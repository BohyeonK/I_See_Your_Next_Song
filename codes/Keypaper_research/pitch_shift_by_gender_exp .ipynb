```python
import bisect
import csv
import functools
import os
import random

import numpy as np
import pandas as pd

import torch
import torchaudio
torchaudio.set_audio_backend('soundfile')

from torch.nn import functional as F
from torch.utils.data import Dataset
from torchaudio import sox_effects, transforms
```

```python
sample_rate = 16000 
input_len = 128000
fft = transforms.Spectrogram(n_fft=800)

data = pd.read_csv('data/all.csv')
files = tuple(data['vocal'])
labels = tuple(data['label'])
gender = 1

uniq_labels = sorted(set(labels))
num_classes = len(uniq_labels)

labels = [bisect.bisect_left(uniq_labels, label) for label in labels]

def load(index):
    audio, sample_rate = torchaudio.load(files[index])
    if sample_rate != 16000:
        transform = transforms.Resample(sample_rate, 16000)
        audio = transform(audio)
        sample_rate = 16000
    assert sample_rate == sample_rate
    return torch.unsqueeze(torch.mean(audio, axis=0), dim=0)  # make it mono

def reshape(audio, length):
    current = audio.shape[1]
    if current < length:
        audio = F.pad(audio, (0, length - current))
    elif current > length:
        idx = random.randint(0, current - length)
        audio = audio[:, idx: idx + length]
    return audio

def pitch_shift_by_male(audio):
    source = reshape(audio, input_len + 50)
    pitch = 6 #random.choice([0,6])
    effects = [['pitch', str(pitch * 100)], ['rate', str(16000)]]
    target, sample_rate = sox_effects.apply_effects_tensor(source, 16000, effects)

    assert sample_rate == sample_rate

    return reshape(target, input_len)

def pitch_shift_by_female(audio):
    source = reshape(audio, input_len + 50)
    pitch = -6 #random.choice([-6,0])
    effects = [['pitch', str(pitch * 100)], ['rate', str(16000)]]
    target, sample_rate = sox_effects.apply_effects_tensor(source, 16000, effects)

    assert sample_rate == sample_rate

    return reshape(target, input_len)
```
