```python
def make_min2_mel(song_paths, save_dir, sample_rate=16000): # song_paths = dataset1
    
    for j in tqdm(range(len(song_paths))):
        wav, wav_sample_rate = torchaudio.load(song_paths[j])
          
        if wav_sample_rate != sample_rate:
            transform = torchaudio.transforms.Resample(wav_sample_rate, sample_rate)
            wav = transform(wav)
                

        wav = torch.unsqueeze(torch.mean(wav, axis=0), dim=0) # mono
        wav_to_mel = torchaudio.transforms.MelSpectrogram(sample_rate=16000, n_fft=512, hop_length=256, n_mels=48) # to mel

        mel = wav_to_mel(wav)
        mel = mel.cpu().detach().numpy() # to numpy
        
        song_name = os.path.splitext(song_paths[j])[0].split('/')[-1]
        np.save(f"{save_dir}/{song_name}.npy", mel)
```

```python
datasets = [dataset1,dataset2,dataset3,dataset4,dataset5,dataset6,dataset7,dataset8]

for i,dataset in enumerate(datasets):
    save_dir = f"{base_dir}/mel_dataset{i+1}"   
    make_min2_mel(song_paths = dataset,save_dir = save_dir)
```
